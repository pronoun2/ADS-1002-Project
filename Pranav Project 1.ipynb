```python
import pandas as pd
from glob import glob
import seaborn as sns
```

```python
demand_raw = sorted(glob('ALL_DEMAND_DATA/*.csv'))

demand = pd.concat((pd.read_csv(file)
           for file in demand_raw), ignore_index = True)
demand
```

```python
# All states data
demand = demand.rename({'SETTLEMENTDATE': 'Date'}, axis=1) 
demand = demand.drop(["PERIODTYPE"], axis = 1)
demand['Date']= pd.to_datetime(demand['Date'])
demand.set_index('Date', inplace=True)
demand
```

```python
# My state data
demand_sa = demand[demand["REGION"]=='SA1']
demand_sa = demand_sa.drop(["REGION"], axis = 1)
demand_sa.rename({'TOTALDEMAND': 'Total Demand'}, axis=1, inplace = True)
```

```python
demand_sa['year'] = demand_sa.index.year
demand_sa['month'] = demand_sa.index.month
demand_sa['day'] = demand_sa.index.day
demand_sa['hour'] = demand_sa.index.hour
demand_sa['minute'] = demand_sa.index.minute

Season = []

for i in range(len(demand_sa)):
    if demand_sa['month'][i] in [12,1,2]:
        Season.append("Summer")
    elif demand_sa['month'][i] in [3,4,5]:
        Season.append("Autumn")
    elif demand_sa['month'][i] in [6,7,8]:
        Season.append("Winter")
    elif demand_sa['month'][i] in [9,10,11]:
        Season.append("Spring")

demand_sa['Season'] = Season

demand_sa = demand_sa.loc['2000-01-01 01:30:00':'2019-12-31 23:30:00']

demand_sa
```

```python
temp_files = ['Temperature Data\\HM01X_Data_023090_999999999743964.txt']

temp_raw = pd.concat((pd.read_table(file, sep=",", error_bad_lines=False)
           for file in temp_files), ignore_index = True)
temp_raw
```

```python
# Goal: Checking to see which columns are uesless so we can drop them.

# Checking if the "Quality" columns are useful. 
for i in temp_raw.columns:
    strings = i.split()
    if strings[0] == 'Quality':
        print(temp_raw[i].unique())

# Checking to see if "AWS Flag", "#" and "hm" are useless.
print(temp_raw['#'].unique())      
print(temp_raw['hm'].unique())
```

```python
# Goal: Drop useless columns and rename columns to something nicer.

temp_cleaned = temp_raw.drop( ['hm',
                               'Year Month Day Hour Minutes in YYYY.1', 'MM.1', 'DD.1', 'HH24.1',
                               'MI format in Local standard time',
                               'Quality of precipitation since 9am local time',
                               'Quality of relative humidity',
                               'Wind speed quality', 
                               'Wind direction quality',
                               'Quality of speed of maximum windgust in last 10 minutes',
                               'Quality of mean sea level pressure',
                               'Quality of station level pressure',
                               'AWS Flag', # Just tells us how the measreument was taken (automatic, manned or hybrid)
                               '#', 
                               'Quality of air temperature',
                               'Quality of Wet bulb temperature',
                               'Quality of dew point temperature'], axis = 1)

temp_sa = temp_cleaned.rename({'Year Month Day Hour Minutes in YYYY': 'year', 
                                'MM':'month', 'DD':'day', 'HH24':'hour', 
                                'MI format in Local time':'minute',
                                'Precipitation since 9am local time in mm':'Percipitation (mm)',
                                'Air Temperature in degrees C': 'Air temperature (C)',
                                'Wet bulb temperature in degrees C': 'Wet Bulb temperature (C)',
                                'Dew point temperature in degrees C': 'Dew Point temperature (C)',
                                'Relative humidity in percentage %': 'Relative humidity (%)',
                                'Wind speed in km/h':'Wind speed (km/h)',
                                'Wind direction in degrees true': 'Wind direction (degrees)',
                                'Speed of maximum windgust in last 10 minutes in  km/h':'Max wind speed (km/h)',
                                'Mean sea level pressure in hPa': 'Sea level pressure (hPa)',
                                'Station level pressure in hPa': 'Station level pressure (hPa)'}, axis=1) 

temp_sa['Date'] = pd.to_datetime(temp_sa[['year', 'month', 'day', 'hour', 'minute']])
temp_sa = temp_sa.set_index('Date')
temp_sa
```

```python
Season = []

for i in range(len(temp_sa)):
    if temp_sa['month'][i] in [12,1,2]:
        Season.append("Summer")
    elif temp_sa['month'][i] in [3,4,5]:
        Season.append("Autumn")
    elif temp_sa['month'][i] in [6,7,8]:
        Season.append("Winter")
    elif temp_sa['month'][i] in [9,10,11]:
        Season.append("Spring")

temp_sa['Season'] = Season

temp_sa
```

```python
# Goal: Find how many missing values per variable and convert everything into strings.
n = len(temp_sa)

columns = []
for column in temp_sa.columns:
    if column not in ['year', 'month', 'day', 'hour', 'minute', 'Season']:
        columns.append(column)

for c in columns:
    missing = 0
    temp_sa[c] = temp_sa[c].astype(str)
    for element in temp_sa[c]:
        if element.strip() == '':
            missing += 1
    print(c, " has", missing, " missing values.")
```

```python
temp_sa2 = temp_sa.copy()

# Goal: Take out rows with empty strings and convert numeric variables to floats.

columns = []
for column in temp_sa2.columns:
    if column not in ['year', 'month', 'day', 'hour', 'minute', 'Season']:
        columns.append(column)

for column in columns:
    temp_sa2 = temp_sa2[[element.strip() != '' for element in temp_sa2[column]]]
    temp_sa2[column] = pd.to_numeric(temp_sa2[column], errors = 'raise')

temp_sa2
```

```python
# Goal: Join temp_vic2 and demand_vic

# Remove one of the rows with duplicates columns (keep first occurrence):
temp_sa2 = temp_sa2[~temp_sa2.index.duplicated(keep='first')]
demand_sa = demand_sa[~demand_sa.index.duplicated(keep='first')]

# Join dataframes
sa_all = pd.concat([demand_sa.drop({'year', 'month', 'day', 'hour', 'minute','Season'}, axis = 1),
                 temp_sa2.drop({'year', 'month', 'day', 'hour', 'minute'}, axis = 1)],
                 axis=1, join="inner")

sa_all
```
